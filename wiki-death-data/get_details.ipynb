```python
# add timestamp of death
from datetime import date
from datetime import datetime

def mtn(x):
    months = {
        'January': '01',
        'February': '02',
        'March': '03',
        'April':'04',
         'May':'05',
         'June':'06',
         'July':'07',
         'August':'08',
         'September':'09',
         'October':'10',
         'November':'11',
         'December':'12'
        }
    try:
        return months[x]
    except:
        raise ValueError('Not a month')

file = pd.read_csv('E:/the economist/wiki_death/data/output/people.csv', encoding='utf-16')
file['timestamp_of_death'] = None


for i in range(122):
    date = file.loc[i,'date_of_death']
    date_sp = date.split()
    month = mtn(date_sp[0])
    day = date_sp[1]
    day = "{:02d}".format(int(day))
    year = file.loc[i,'year_of_death']
    file.loc[i,'timestamp_of_death'] = str(year) + month + str(day)


d = file.pop('timestamp_of_death')
file.insert(7,'timestamp_of_death',d) 
file = file.drop(['Unnamed: 0','Unnamed: 0.1'],axis = 1)
file.to_csv('E:/the economist/wiki_death/data/output/people.csv', encoding='utf-16')
```

```python
# add median_views_before and median_views_after
file = pd.read_csv('E:/the economist/wiki_death/data/output/people.csv', encoding='utf-16')
path = 'E:/the economist/wiki_death/data/cele_join'
file['median_views_before'] = None
file['median_views_after'] = None
for i in range(122):
    f_name =  file.loc[i,'id']
    file2 = pd.read_csv(path + '/' + f_name + '.csv', encoding = 'utf-16')
    timestamp_of_death = file.loc[i,'timestamp_of_death']
    timestamp_of_death = str(timestamp_of_death) + '00'
    timestamp_of_death = numpy.int64(timestamp_of_death)
    views_before = file2[file2['timestamp'] < timestamp_of_death]
    views_after = file2[file2['timestamp'] > timestamp_of_death]
    median_views_before = views_before['views'].median()
    median_views_after = views_after['views'].median()
    file.loc[i, 'median_views_before'] = median_views_before
    file.loc[i,'median_views_after'] = median_views_after

d = file.pop('median_views_before')
file.insert(10,'median_views_before',d)
d = file.pop('median_views_after')
file.insert(11,'median_views_after',d)
file = file.drop(['Unnamed: 0'],axis = 1)
file.to_csv('E:/the economist/wiki_death/data/output/people.csv', encoding='utf-16')
```

```python
# add median_views_adjusted_before & max_views_adjusted & max_change_baseline_views_adjusted
file = pd.read_csv('E:/the economist/wiki_death/data/output/people.csv', encoding='utf-16')
path = 'E:/the economist/wiki_death/data/cele_join'
file['median_views_adjusted_before'] = None
file['max_views_adjusted'] = None
file['max_change_baseline_views_adjusted'] = None
for i in range(122):
    f_name =  file.loc[i,'id']
    file2 = pd.read_csv(path + '/' + f_name + '.csv', encoding = 'utf-16')
    timestamp_of_death = file.loc[i,'timestamp_of_death']
    timestamp_of_death = numpy.int64(timestamp_of_death)
    views_adjusted_before = file2[file2['timestamp'] < timestamp_of_death]
    median_views_adjusted_before = views_adjusted_before['views_adjusted'].median()
    max_views_adjusted = file2['views_adjusted'].max()
    max_change_baseline_views_adjusted = max_views_adjusted / median_views_adjusted_before
    file.loc[i, 'median_views_adjusted_before'] = median_views_adjusted_before
    file.loc[i,'max_views_adjusted'] = max_views_adjusted
    file.loc[i,'max_change_baseline_views_adjusted'] = max_change_baseline_views_adjusted

d = file.pop('median_views_adjusted_before')
file.insert(10,'median_views_adjusted_before',d)
d = file.pop('max_views_adjusted')
file.insert(11,'max_views_adjusted',d)
d = file.pop('max_change_baseline_views_adjusted')
file.insert(12,'max_change_baseline_views_adjusted',d)
file = file.drop(['Unnamed: 0'],axis = 1)
file.to_csv('E:/the economist/wiki_death/data/output/people.csv', encoding='utf-16')
```

```python
# add median_percent_traffic_before and median_percent_traffic_after
file = pd.read_csv('E:/the economist/wiki_death/data/output/people.csv', encoding='utf-16')
path = 'E:/the economist/wiki_death/data/cele_join'
file['median_percent_traffic_before'] = None
file['median_percent_traffic_after'] = None
for i in range(122):
    f_name =  file.loc[i,'id']
    file2 = pd.read_csv(path + '/' + f_name + '.csv', encoding = 'utf-16')
    timestamp_of_death = file.loc[i,'timestamp_of_death']
    timestamp_of_death = numpy.int64(timestamp_of_death)
    traffic_before = file2[file2['timestamp'] < timestamp_of_death]
    traffic_after = file2[file2['timestamp'] > timestamp_of_death]
    median_percent_traffic_before = traffic_before['percent_traffic'].median()
    median_percent_traffic_after = traffic_after['percent_traffic'].median()
    file.loc[i, 'median_percent_traffic_before'] = median_percent_traffic_before
    file.loc[i,'median_percent_traffic_after'] = median_percent_traffic_after

d = file.pop('median_percent_traffic_before')
file.insert(13,'median_percent_traffic_before',d)
d = file.pop('median_percent_traffic_after')
file.insert(14,'median_percent_traffic_after',d)
file['median_percent_traffics'] = ['%.9f' % x for x in file['median_percent_traffics'].values]
file['median_percent_traffic_before'] = ['%.9f' % x for x in file['median_percent_traffic_before'].values]
file['median_percent_traffic_after'] = ['%.9f' % x for x in file['median_percent_traffic_after'].values]
file = file.drop(['Unnamed: 0'],axis = 1)
file.to_csv('E:/the economist/wiki_death/data/output/people.csv', encoding='utf-16')
```

```python
# add std & iqr
file = pd.read_csv('E:/the economist/wiki_death/data/output/people.csv', encoding='utf-16')
path = 'E:/the economist/wiki_death/data/cele_join'
file['std'] = None
file['iqr'] = None
file['mean_views_adjusted_before'] = None

for i in range(122):
    f_name =  file.loc[i,'id']
    file2 = pd.read_csv(path + '/' + f_name + '.csv', encoding = 'utf-16')
    timestamp_of_death = file.loc[i,'timestamp_of_death']
    timestamp_of_death = numpy.int64(timestamp_of_death)
    traffic_before = file2[file2['timestamp'] < timestamp_of_death]
    std = traffic_before['views_adjusted'].std()
    iqr = traffic_before['views_adjusted'].quantile(0.75) - traffic_before['views_adjusted'].quantile(0.25)
    mean_views_adjusted_before = traffic_before['views_adjusted'].mean()
    file.loc[i, 'std'] = std
    file.loc[i,'iqr'] = iqr
    file.loc[i,'mean_views_adjusted_before'] = mean_views_adjusted_before
    

d = file.pop('std')
file.insert(13,'std',d)
d = file.pop('iqr')
file.insert(14,'iqr',d)
d = file.pop('mean_views_adjusted_before')
file.insert(12,'mean_views_adjusted_before',d)
file = file.drop(['Unnamed: 0'],axis = 1)
file.to_csv('E:/the economist/wiki_death/data/output/people.csv', encoding='utf-16')
```
