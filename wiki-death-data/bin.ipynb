## Bin by week, death day is at the first of the week

```python
import pandas as pd
import os
```

```python
# add time_index for total wiki views file
file = pd.read_csv('E:/the economist/wiki_death/data/wiki_pageviews.csv', encoding = 'utf-8')
file['timestamp'] = file['timestamp'].astype(str)
file['timestamp'] = file.timestamp.map(lambda x: x[:-2])
file['timeIndex'] = None
timestamp = list(set(file['timestamp'].values.tolist()))
timestamp.sort()
for i in range(len(timestamp)):
    file.loc[file['timestamp']==timestamp[i], 'timeIndex'] = i

file = file.drop(['Unnamed: 0'], axis = 1)
d = file.pop('timeIndex')
file.insert(1,'timeIndex',d) 
file.to_csv('E:/the economist/wiki_death/data/wiki_pageviews.csv', encoding = 'utf-8')
```

```python
# bin by week
file = pd.read_csv('E:/the economist/wiki_death/data/wiki_pageviews.csv', encoding = 'utf-8')
path = 'E:/the economist/wiki_death/data/cele_join'
lst = os.listdir(path)
h = 1
for c in lst:
    file_c = pd.read_csv(path + '/' + c, encoding = 'utf-16')
    file_c['bin_death_index'] = None
    file_c['rem'] = None
    for j in range(len(file_c)):
        file_c.loc[j,'bin_death_index'] = file_c.loc[j,'death_index'] // 2 #
        file_c.loc[j,'rem'] = file_c.loc[j,'death_index'] % 2 #
    
    num = list(set(file_c['bin_death_index'].values.tolist()))
    num.sort()
    file_l = pd.DataFrame() 
    file_l['bin'] = None
    file_l['timestamp_index'] = None
    file_l['bin_death_index'] = None
    file_l['timestamp'] = None
    file_l['views'] = None
    file_l['percent_traffic'] = None
    file_l['views_adjusted'] = None
    for i in num:
        df = file_c[file_c['bin_death_index'] == i]
        ind = num.index(i)
        if len(df) == 2: #
            file_l.loc[ind, 'bin'] = ind
            file_l.loc[ind,'timestamp_index'] = df.loc[df['rem'] == 0, 'timeIndex'].values[0]
            file_l.loc[ind,'bin_death_index'] = i
            file_l.loc[ind,'timestamp'] = df.loc[df['rem'] == 0, 'timestamp'].values[0]
            file_l.loc[ind,'views'] = df['views'].sum()
            file_l.loc[ind,'views_adjusted'] = df['views_adjusted'].sum()
            timeIndex = df.loc[df['rem'] == 0, 'timeIndex'].values[0]
            file2 = file[file['timeIndex'].isin(range(timeIndex,timeIndex + 2))] #
            file_l.loc[ind,'percent_traffic'] = file_l.loc[ind,'views'] / file2['views'].sum()
     
    file_l['pageid'] = file_c.loc[0, 'pageid']
    d = file_l.pop('pageid')
    file_l.insert(2,'pageid',d) 
    file_l.to_csv('E:/the economist/wiki_death/data/cele_join_2' + '/' + c, encoding = 'utf-8') #
    print(h,'/122')
    h = h+1


# contents.to_csv('E:/the economist/wiki_death/data/explore.csv', encoding='utf-8')
```
