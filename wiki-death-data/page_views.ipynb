# 爬每个名人的页面浏览量

```python
s = open("E:/wiki_death/data/people_data.csv","r",encoding="utf-16") 
i = 0
info = csv.reader(s)  

for row in info:
    if i>926:
        link = row[2]
        name = link[30:]
        print(name)
        print(link)
        pageview = pageviewapi.per_article('en.wikipedia.org', name, '2018010100', '2023010100',
                            access='all-access', agent='user', granularity='daily')
        contents = pd.DataFrame(pageview['items'])
        contents.to_csv("E:/wiki_death/data/celebrities/" + name + ".csv", encoding='utf-8')
    i = i + 1
```
